```python
import numpy as np
import pandas as pd
import timeit
```

```python
df = pd.read_csv("C:\\Users\\Alican\\03-Python Data Science Bootcamp\\tips.csv")
```

```python
df.info()
```

```python
# apply() method applies the non-builtin funcitons to dataframe
def last_four_digits(num):
    
    return int(str(num)[-4:])
```

```python
df['last_four'] = df['CC Number'].apply(last_four_digits)
df.head(10)
```

```python
def price_category(num):
    
    num = int(num)
    
    if num in range(0,11):
        
        return '$'
    
    elif num in range(11,22):
        
        return '$$'
    
    elif num in range(22,33):
        
        return '$$$'
    
    else:
        
        return '$$$$'   
```

```python
df['price_category'] = df['total_bill'].apply(price_category)
df.head(10)
```

```python
def tip_quality(total_bill, tip):
    
    tip_pct = np.round(100 * tip / total_bill, 2)
    
    if tip_pct > 15.00:
        
        return 'Generous' + f'(Pct:%{str(tip_pct)})'
    
    else:
        
        return 'Poor' + f'(Pct:%{str(tip_pct)})'
   
```

```python
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df: tip_quality(df['total_bill'], df['tip']), axis = 1)
df.head(10)
```

```python
# Both do same thing
'''
np.vectorize is just a convenience function. 
It doesn't actually make code run any faster.

The purpose of np.vectorize is to transform functions which are not numpy-aware 
(e.g. take floats as input and return floats as output) into functions
that can operate on (and return) numpy arrays.
''' 
```

```python
%%timeit
df['tip_quality'] = df[['total_bill', 'tip']].apply(lambda df: tip_quality(df['total_bill'], df['tip']), axis = 1)
df.head(10)
```

```python
%%timeit
df['tip_quality'] = np.vectorize(tip_quality)(df['total_bill'], df['tip'])
df.head(10)
```
