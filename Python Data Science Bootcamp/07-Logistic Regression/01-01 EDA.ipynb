```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('hearing_test.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(df['test_result']);
df['test_result'].value_counts()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='test_result', y='age', data=df, width=0.5);
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='test_result', y='physical_score', data=df, width=0.5);
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='age', y='physical_score', data=df, hue='test_result', alpha=0.5);
```

```python
# We can smimply do this instead of we did at the top
sns.pairplot(data=df, hue='test_result', height=6);
```

```python
sns.heatmap(data=df.corr(), annot=True);
```

```python
#3D Data
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(
    df['age'], 
    df['physical_score'], 
    df['test_result'], 
    c=df['test_result']
);
```
