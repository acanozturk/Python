```python
l = df.columns.tolist()[:-1]
categorical_cols = []
for i in range(len(l)):
    print(f'{l[i]}:{df[l[i]].unique()}')
    
    if len(df[l[i]].unique()) <= 5:
        categorical_cols.append(l[i])
```

```python
for c in categorical_cols:
    df[c] = df[c].apply(str)
```

```python
cat_cols = df.select_dtypes(include = 'object')
num_cols = df.select_dtypes(exclude = 'object')
```

```python
df_dummies = pd.get_dummies(cat_cols, drop_first = True)
final_df = pd.concat([num_cols, df_dummies], axis = 1)
```

```python
final_df.to_csv('heart_eda.csv', index = False)
```
