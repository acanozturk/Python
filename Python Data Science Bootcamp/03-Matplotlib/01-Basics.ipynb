```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
# They all should be on same cell
# Basic plotting
plt.plot(x, y);
plt.title('y = 2x');
plt.xlabel('x');
plt.xlim(0,6);
plt.ylabel('y');
plt.ylim(0,15);
plt.savefig('exampleplot.png');
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
# Figures
fig = plt.figure(figsize = (10, 5))
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlim(0,10);
axes1.set_ylim(0,10000);
axes1.set_xlabel('a');
axes1.set_ylabel('b');
axes1.set_title('b = a^4');

axes2 = fig.add_axes([0.1,0.25,0.5,0.5])
axes2.plot(x,y);
axes2.set_xlim(0,10);
axes2.set_ylim(0,20);
axes2.set_xlabel('x');
axes2.set_ylabel('y');
axes2.set_title('y = 2x');
```

```python
# Sub-plots
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[(0,0)].plot(x,y)
axes[(0,0)].set_title('y = 2x')

axes[(0,1)].plot(y,x**2)
axes[(0,1)].set_title('y = x^2')

axes[(1,0)].plot(a,b)
axes[(1,0)].set_title('b = a^4')

axes[(1,1)].plot(b,a)
axes[(1,1)].set_title('a = b^0.25')

plt.tight_layout(h_pad = 2, w_pad = 2) #
```

```python
# Legend
fig = plt.figure(figsize = (5, 5))
axes = fig.add_axes([0,0,1,1])
axes.plot(a, label = 'a');
axes.plot(a**2, label = 'a^2');
axes.legend(loc = 9)
```

```python
# Colors and Styles
fig = plt.figure(figsize = (10, 5))
axes = fig.add_axes([0,0,1,1])
axes.plot(a, color = 'purple', lw = 3, ls = '--', marker = 's', markersize = 10, markerfacecolor = 'orange', markeredgewidth = 4, markeredgecolor = 'cyan');
axes.plot(a**2, color = 'red', lw = 2, ls = '-.', marker = 'x', ms = 10);
axes.plot(a**2.2, color = 'brown', lw = 1, ls = ':', marker = 'o', ms = 10);
```
