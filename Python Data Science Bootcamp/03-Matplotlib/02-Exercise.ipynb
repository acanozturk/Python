```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1
csqr =  299792458 ** 2
m = np.linspace(0, 10, 11)
e = m * csqr

plt.plot(e, lw = 5, color = 'red')
plt.xlabel('mc^2')
plt.ylabel('e')
plt.title('e = mc^2')
plt.grid(True)
```
