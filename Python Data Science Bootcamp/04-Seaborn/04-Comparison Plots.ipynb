```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head() 
```

```python
# Joint plot
sns.jointplot(data = df, 
              x = 'math score', 
              y = 'reading score',
              hue = 'gender');
```

```python
# Pair plot
# For numeric values, usually called after filtering data
sns.pairplot(data = df,
             hue = 'gender');
```
