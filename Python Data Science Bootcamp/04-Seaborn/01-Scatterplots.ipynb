```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
df.shape
```

```python
# Since seaborn based on matpoltlib, we can apply it's operations
plt.figure(figsize = (18, 5), dpi = 200)
sns.scatterplot(x = 'salary', 
                y = 'sales',
                data = df, 
                hue = 'division', 
                palette = 'magma', # Color palette
                alpha = 0.8, # Transparency
                style = 'division' # Should be categorical
               );
```
