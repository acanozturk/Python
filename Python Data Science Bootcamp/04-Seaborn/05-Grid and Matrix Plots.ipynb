```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head() 
```

```python
# Cat plot
sns.catplot(data = df, 
            x = 'math score', 
            y = 'gender', 
            kind = 'box',
            row = 'test preparation course',
            col = 'lunch');
```
