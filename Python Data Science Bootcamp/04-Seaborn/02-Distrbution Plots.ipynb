```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
# Rug plot
sns.rugplot(x = 'salary', data = df, height = 0.5);
```

```python
# Displot is generalized version of histplot
sns.set(style = 'darkgrid')
sns.displot(data = df, 
            x = 'salary', 
            bins = 25, 
            color = 'red',
            edgecolor = 'blue', 
            kde = True); # Denstiy curve
```

```python
sns.histplot(data = df, 
             x = 'salary')
```

```python
np.random.seed(42)
ages = pd.DataFrame(np.random.randint(0, 100, 300), columns = ['age'])
sns.displot(data = ages, 
            x = 'age', 
            color = 'purple',
            bins = 30, 
            kde = True,);
```

```python
sns.kdeplot(data = ages, x = 'age');
```
