```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('application_record.csv')
df.head().T
```

```python
df.info()
```

```python
employeed = df[df['DAYS_EMPLOYED'] < 0]
employeed
```

```python
# 1
employeed['DAYS_BIRTH'] = -1 * employeed['DAYS_BIRTH']
employeed['DAYS_EMPLOYED'] = -1 * employeed['DAYS_EMPLOYED']
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = 'DAYS_BIRTH', 
                y = 'DAYS_EMPLOYED',
                data = employeed,  
                alpha = 0.01);
```

```python
# 2
df['AGE_YEARS'] = np.round(np.abs(df['DAYS_BIRTH']) / 365).astype(int)
```

```python
df['AGE_YEARS']
```

```python
sns.displot(data = df,
            x = 'AGE_YEARS',
            color = 'pink',
            bins = 45);
```

```python
# 3
lh = df[df['AMT_INCOME_TOTAL'] < df['AMT_INCOME_TOTAL'].mean()]
lh
```

```python
plt.figure(figsize = (12, 8));
sns.boxplot(data = lh, 
            x = 'NAME_FAMILY_STATUS', 
            y = 'AMT_INCOME_TOTAL', 
            hue = 'FLAG_OWN_REALTY');
plt.legend(bbox_to_anchor = (1,1));
```
