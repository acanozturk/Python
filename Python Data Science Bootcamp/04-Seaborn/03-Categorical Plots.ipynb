```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df['division'].value_counts()
```

```python
# Count plot
plt.figure(figsize = (10,5))
sns.countplot(data = df, 
              x = 'division', 
              hue = 'level of education',
              palette = 'Set2');
```

```python
# Bar plot
plt.figure(figsize = (10,5), dpi = 150)
sns.barplot(data = df, 
            x = 'level of education', 
            y = 'sales', 
            hue = 'division',
            estimator = np.mean);
plt.legend(bbox_to_anchor = (1, 1));
```

```python
# Box plot
plt.figure(figsize = (10,5), dpi = 150)
sns.boxplot(data = df, 
            x = 'salary', 
            y = 'level of education')
```

```python
df2 = pd.read_csv('StudentsPerformance.csv')
df2.head()
```

```python
plt.figure(figsize = (10,5), dpi = 150)
sns.boxplot(data = df2,
            x = 'math score',
            y = 'test preparation course',
            hue = 'parental level of education',
            palette = 'Set3');
plt.legend(bbox_to_anchor = (1, 1));
```

```python
# Violin plot
plt.figure(figsize = (10,5), dpi = 150)
sns.violinplot(data = df2,
               x = 'math score',
               y = 'test preparation course',
               hue = 'parental level of education',
               palette = 'Set3');
plt.legend(bbox_to_anchor = (1, 1));
```

```python
# Swarm plot
plt.figure(figsize = (10,5), dpi = 150)
sns.swarmplot(data = df2,
               x = 'math score',
               y = 'test preparation course');
```
