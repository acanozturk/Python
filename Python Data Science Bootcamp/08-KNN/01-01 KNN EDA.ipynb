```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('gene_expression.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.pairplot(
    data=df,
    hue='Cancer Present',
    height=6
);
```

```python
methods = {'pearson', 'kendall', 'spearman'}
for method in methods:
    print(f'Method: {method.capitalize()}')
    print(df.corr(method=method)['Cancer Present'], end='\n'*2)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    data=df.corr(),
    annot=True
);
```
